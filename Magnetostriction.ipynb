```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def posicion_letra(letra):
    return ord(letra.lower()) - ord('a')

# Cargar el archivo Excel
df = pd.read_excel('/content/Book1-1 (1).xlsx')

# Seleccionar los datos desde la fila 3 hasta la 40 (índices 2 a 39, porque Python empieza en 0)
# Y seleccionar las columnas C y D (índices 2 y 3 si las columnas están como A=0, B=1, C=2, D=3)
x = df.iloc[1:19, posicion_letra("R")].tolist()  # Columna R
y = df.iloc[1:19, posicion_letra("Q")].tolist()  # Columna Q

# Mostrar un vistazo a los datos
print("x:", x[:5])  # Muestra los primeros 5 elementos de x
print("y:", y[:5])  # Muestra los primeros 5 elementos de y

plt.scatter(x, y)
plt.xlabel(f'Campo magnético B [mT]')
plt.ylabel(r'$\Delta\ell\hspace{0.3}/\hspace{0.3}\ell$')
plt.title('Niquel')
plt.grid(True)
plt.show()

```
